```python
import numpy as np
import symlib
from mwgcs import SymphonyHaloProfile, NFW, Einasto
from gala.potential import NFWPotential
from gala.units import galactic
import matplotlib.pyplot as plt
import astropy.units as u
import asdf
from tqdm import tqdm
from scipy.optimize import curve_fit
import glob
import re
import gravitree

sim_dirs = glob.glob('/fs/ddn/sdf/group/kipac/g/cosmo/ki21/phil1/simulations/ZoomIns/SymphonyMilkyWay/Halo*')
```

```python
profiles_dir = "/fs/ddn/sdf/group/kipac/g/cosmo/ki21/phil1/data/subhalo_profiles/*z0_b.npy"

# checking against phil's profiles
res = [f for f in glob.glob(profiles_dir) if re.search(r"/fs/ddn/sdf/group/kipac/g/cosmo/ki21/phil1/data/subhalo_profiles/prof_SymphonyMilkyWay_[A-Za-z0-9]+_z0_b\.npy", f)]

bin_edges = np.load("/fs/ddn/sdf/group/kipac/g/cosmo/ki21/phil1/data/subhalo_profiles/r_bins.npy")

fig, ax = plt.subplots(dpi=200, figsize=(6,6))

for f in res:

    n = np.cumsum(np.load(f)[0, :])

    r = bin_edges[1:]
    
    m = np.asarray(n, dtype=float)

    _accelerations = m / (r**2)
    
    idx_0 = np.argmax(_accelerations)
    
    a_0 = _accelerations[idx_0]
    r_0 = r[idx_0]
    
    ax.plot(r/r_0, _accelerations, c='r', alpha=.25)

ax.set_xscale('log')
ax.set_xlabel(r'$r/r_0$')
ax.set_ylabel(r'$a(<r)/a_0$')

plt.show()
```

```python
profile_acceleration = []

bound = True

for sim_dir in tqdm(sim_dirs):

    # hosts only
    halo_id = 0

    snapshot = 235 
    
    rs, rs_hist = symlib.read_rockstar(sim_dir)
    sf, hist = symlib.read_symfind(sim_dir)
    
    params = symlib.simulation_parameters(sim_dir)
    scale_factors = symlib.scale_factors(sim_dir)
    mp = params['mp'] / params['h100']
    particle_class = symlib.Particles(sim_dir)
    particles = particle_class.read(snapshot, mode='all')

    rvir = rs[halo_id, snapshot]['rvir']
    
    q = particles[halo_id]['x']
    q_sh = sf[halo_id, snapshot]['x']
    p = particles[halo_id]['v']

    dq = q - q_sh
    dp = p - sf["v"][halo_id, snapshot]
    
    r = np.sqrt(np.sum(dq**2, axis=1)) / rvir
    order = np.argsort(r)

    ke = np.sum(dp**2, axis=1) / 2
    ok = np.ones(len(ke), dtype=bool)

    for _ in range(3):
        _, vmax, pe, _ = symlib.profile_info(params, dq, ok=ok)
        E = ke + pe*vmax**2
        ok = E < 0
    
    # profile = SymphonyHaloProfile(q[ok], q_sh, mp, rvir)
    softening = params['eps']/params['h100'] * scale_factors[snapshot]

    hist = np.histogram(r[ok], bins=bin_edges)[0]
    menc = np.cumsum(hist)
    
    # radii = np.logspace(-3, 2, 1000)
    # menc = np.array([profile.menc(r_i * rvir) for r_i in bin_edges[1:]])
    
    _g_accel = 1.3938323614347172e-22 # ((u.kpc**2 / u.Msun) * (u.km/u.s**2))
    # accel = _g_accel * menc / (radii**2 + softening**2)
    accel = _g_accel * menc / (np.array(bin_edges[1:])**2)

    a_0 = np.max(accel)
    r_0 = bin_edges[1:][np.argmax(accel)]

    profile_acceleration.append(accel / a_0)
    # profile_radii.append(radii / r_0)
```

```python
fig, ax = plt.subplots(dpi=300, figsize=(6,6))
[ax.plot(bin_edges[1:] / bin_edges[1:][np.argmax(a_i)] , a_i, c='r', alpha=.25) for a_i in profile_acceleration]
ax.set_xlabel(r'$r/r_0$')
ax.set_ylabel(r'$a(<r)/a_0$')
ax.set_xscale('log')
fig.suptitle(r'MW Host Acceleration $(z=0)$')
fig.tight_layout()
plt.show()
```
