```python
a_r = (np.array(acc) * u.km / u.s**2).to(u.kpc / u.Myr**2).value

def get_normal_vectors(q_xyz):
    normal = - q_xyz / np.linalg.norm(q_xyz)
    return normal
```

```python
integ_acc = np.zeros((3, len(orbit.t)))

class LeapfrogIntegrator():
    
    def __init__(self, w0, acc, ts):
        self.w0 = w0 # shape is (6,)
        self.acc = acc
        self.ts = ts
        self.steps = len(ts)

        self.w = np.zeros((6, self.steps))
        self.w[:, 0] = self.w0
        self.dt = ts[1] - ts[0] # in Myr

    def loop(self, nsteps):

        q = self.w0[:3]
        p = self.w0[3:]

        self.w[:3, 0] = q
        self.w[3:, 0] = p

        a_0 = acc[0] * get_normal_vectors(q)
        integ_acc[:, 0] = a_0
        # kick
        p_half = p + (self.dt/2. * a_0)
        
        for k in tqdm(range(0, nsteps-1)):

            # drift
            q = q + (self.dt/2. * p_half)
            self.w[:3, k] = q

            # kick
            a_i = acc[k] * get_normal_vectors(q)
            integ_acc[:, k] = a_i
            p_half = p_half + (self.dt * a_i / 2)
            
            self.w[3:, k] = p_half

w0 = orbit.w[:, 0]
# w0[3:] = w0[3:]

integ = LeapfrogIntegrator(orbit.w[:, 0], a_r, orbit.t)
```

```python
integ.loop(len(orbit.t) - 1)
```
