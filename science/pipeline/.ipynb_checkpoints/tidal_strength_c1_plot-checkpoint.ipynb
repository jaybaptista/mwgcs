```python
import numpy as np
import matplotlib.pyplot as plt
from mwgcs import NFW, Einasto
from gala.units import galactic
from gala.potential import PlummerPotential
import astropy.units as u
```

```python
pl_potential = PlummerPotential(1e10, 2. * u.kpc, units=galactic)
nfw_potential = NFW(1e12, 25., 11.)
ein_potential = Einasto(.18, 25., 7.)
```

```python
# make a plot of the acceleration profiles for each
fig, ax = plt.subplots(dpi=200, figsize=(3,3))

r = np.logspace(-2, 2, 100)

pl_acc = np.abs(pl_potential.acceleration([r, 0.*r, 0.*r] * u.kpc)[0, :])
nfw_acc = np.array([np.abs(nfw_potential.acceleration(r_i)) for r_i in r])
ein_acc = np.array([np.abs(ein_potential.acceleration(r_i)) for r_i in r])

# plummer acceleration
ax.plot(r / r[np.argmax(pl_acc)], pl_acc / np.max(pl_acc), c='b', label='plummer')
ax.plot(r / r[np.argmax(nfw_acc)], nfw_acc / np.max(nfw_acc), c='r', label='nfw')
ax.plot(r / r[np.argmax(ein_acc)], ein_acc / np.max(ein_acc), c='orange', ls='--', label='einasto')

ax.set_yscale('log')
ax.set_xscale('log')

ax.legend()

ax.set_xlabel(r'$r/r_0$')
ax.set_ylabel(r'$|a/a_0|$')
plt.show()
```

```python

r = np.logspace(-2, 2, 100)

def hessian_to_lambda(hess, omega=True, ):
    tidal_tensor = - hess
    eigenvalues, eigenvectors = np.linalg.eig(tidal_tensor)

    l1 = eigenvalues[0] 
    l2 = eigenvalues[1]
    l3 = eigenvalues[2] # most negative eigenvalue? 
    _om = - (1/3) * (l1 + l2 + l3)

    # modify omega using eq. 4 and |V x R^2| / R
    if not omega:
        _om = 0.
    return eigenvalues[0] + _om


def getOmega(hess):
    eigenvalues, eigenvectors = np.linalg.eig(tidal_tensor)

    l1 = eigenvalues[0] 
    l2 = eigenvalues[1]
    l3 = eigenvalues[2] # most negative eigenvalue? 
    _om = - (1/3) * (l1 + l2 + l3)
    return _om
    

def getStrength(hess):
    tidal = np.array([hessian_to_lambda(hess[:, :, i]) for i in range(100)])
    no_omega = np.array([hessian_to_lambda(hess[:, :, i], False, True) for i in range(100)])
    test = np.array([hessian_to_lambda(hess[:, :, i], True, True) for i in range(100)])
    return tidal, no_omega, test


pl_hess = pl_potential.hessian([r, 0.*r, 0.*r]).to(1/u.Gyr**2).value
pl_tidal, pl_tidal_no_omega, pl_tidal_test = getStrength(pl_hess)

nfw_hess = np.array([nfw_potential.hessian(r_i) for r_i in r]).T
nfw_tidal, nfw_tidal_no_omega, nfw_tidal_test = getStrength(nfw_hess)

ein_hess = np.array([ein_potential.hessian(r_i) for r_i in r]).T
ein_tidal, ein_tidal_no_omega, ein_tidal_test = getStrength(ein_hess)
```

```python
# make a plot of the tidal profiles for each
fig, ax = plt.subplots(dpi=200, figsize=(4,4))


ax.plot(r, pl_tidal, c='b', label='plummer')
ax.plot(r, nfw_tidal, c='r', label='nfw')
ax.plot(r, ein_tidal, c='orange', label='einasto')

ax.plot(r, pl_tidal_test, c='b', ls='--', alpha=.25)
ax.plot(r, nfw_tidal_test, c='r', ls='--', alpha=.25)
ax.plot(r, ein_tidal_test, c='orange', ls='--', alpha=.25)
ax.plot([], [], c='k', ls='--', alpha=.25, label='trace kept')

ax.plot(r, pl_tidal_no_omega, c='b', ls='-.' )
ax.plot(r, nfw_tidal_no_omega, c='r', ls='-.' )
ax.plot(r, ein_tidal_no_omega, c='orange', ls='-.' )
ax.plot([], [], c='k', ls='-.', label=r'$\lambda_1$')


ax.set_yscale('symlog') # put a curve that does not assume a perfectly circular orbit.
ax.set_xscale('log')

ax.legend()

ax.set_xlabel(r'$r$ [kpc]')
ax.set_ylabel(r'$\frac{\partial^2 \Phi}{\partial r^2} + \Omega^2$')
plt.show()
```

```python
# make a plot of the tidal profiles for each
fig, ax = plt.subplots(dpi=200, figsize=(3,3))


ax.plot(r, pl_tidal_no_omega, c='b', label='plummer')
ax.plot(r, nfw_tidal_no_omega, c='r', label='nfw')
ax.plot(r, ein_tidal_no_omega, c='orange', ls='--', label='einasto')

ax.set_yscale('symlog')
ax.set_xscale('log')

ax.legend()

ax.set_xlabel(r'$r$ [kpc]')
ax.set_ylabel(r'$\lambda_1$')
plt.show()
```
