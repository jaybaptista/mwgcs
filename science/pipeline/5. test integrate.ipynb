```python
def foo(v_sph, q_xyz):
    x, y, z = q_xyz
    v_r, v_theta, v_phi = v_sph
    
    r = np.sqrt(x**2 + y**2 + z**2)
    theta = np.arctan(np.sqrt(x**2 + y**2) / z)
    phi = np.arctan(y / x)

    v_x = (np.sin(theta) * np.cos(phi) * v_r) + (np.cos(theta) * np.cos(phi) * v_theta) - (np.sin(phi) * v_phi)
    v_y = (np.sin(theta) * np.sin(phi) * v_r) + (np.cos(theta) * np.sin(phi) * v_theta) + (np.cos(phi) * v_phi)
    v_z = (np.cos(theta) * v_r) - (np.sin(theta) * v_theta)

    v_xyz = np.array([v_x, v_y, v_z])
    return v_xyz
```

```python
integ_acc = np.zeros((3, len(t.value)))

class LeapfrogIntegrator():
    
    def __init__(self, w0, acc, ts):
        self.w0 = w0 # shape is (6,)
        self.acc = acc
        self.ts = ts
        self.steps = len(ts)

        self.w = np.zeros((6, self.steps))
        self.w[:, 0] = self.w0
        self.dt = ts[1] - ts[0] # in Myr

    def loop(self, nsteps):

        q = self.w0[:3]
        p = self.w0[3:]

        self.w[:3, 0] = q
        self.w[3:, 0] = p

        p_half = p + (self.dt/2. * pot.acceleration(q).value.flatten())

        integ_acc[:, 0] = pot.acceleration(q).value.flatten()
        
        for k in tqdm(range(0, nsteps-1)):
            q = q + (self.dt/2. * p_half)
            self.w[:3, k] = q
            
            p_half = p_half + (self.dt * pot.acceleration(q).value.flatten() / 2)
            integ_acc[:, k] = pot.acceleration(q).value.flatten()
            self.w[3:, k] = p_half

w0 =  w0_tmp

# acc[2, :] = 0

integ = LeapfrogIntegrator(w0, acc, t.value)
```
