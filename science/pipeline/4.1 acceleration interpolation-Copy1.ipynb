```python
acc_sph = np.array([-acc, 0.*acc, 0.*acc])
```

```python
def foo(v_sph, q_xyz):
    x, y, z = q_xyz
    v_r, v_theta, v_phi = v_sph
    
    r = np.sqrt(x**2 + y**2 + z**2)
    theta = np.arctan(np.sqrt(x**2 + y**2) / z)
    phi = np.arctan(y / x)

    v_x = (np.sin(theta) * np.cos(phi) * v_r) + (np.cos(theta) * np.cos(phi) * v_theta) - (np.sin(phi) * v_phi)
    v_y = (np.sin(theta) * np.sin(phi) * v_r) + (np.cos(theta) * np.sin(phi) * v_theta) + (np.cos(phi) * v_phi)
    v_z = (np.cos(theta) * v_r) - (np.sin(theta) * v_theta)

    v_xyz = np.array([v_x, v_y, v_z])
    return v_xyz

acc_xyz = foo(acc_sph, w[:3, :-1])
```

```python
acc_true = (acc_xyz * u.km / u.s**2).to(u.kpc / u.Myr**2).value
```

```python
def get_normal_vectors(q_xyz):
    normal = - q_xyz / np.linalg.norm(q_xyz)
    return normal

norm_xyz = get_normal_vectors(w[:3])
```

```python
fig, ax = plt.subplots(dpi=200)

x, y = w[0, :], w[1, :]
a_x, a_y = acc_true[0, :], acc_true[1, :]
n_x, n_y = norm_xyz[0, :], norm_xyz[1, :]
ax.plot(x, y, alpha=.2)
k = 10
# ax.quiver(x[::k], y[::k], a_x[::k]*1e2, a_y[::k]*1e2, angles='xy', scale_units='xy', scale=.01, color='b')

ax.quiver(x[::k], y[::k], n_x[::k], n_y[::k], angles='xy', scale_units='xy', scale=.001, color='b')

ax.set_xlabel('x [kpc]')
ax.set_ylabel('y [kpc]')

ax.set_aspect('equal')
plt.show()
```
