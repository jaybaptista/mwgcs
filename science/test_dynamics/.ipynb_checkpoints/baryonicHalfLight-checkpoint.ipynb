```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def rh_rvir_relation(rvir, addScatter=True):
    slope = .95
    normalization = .015
    scatter = 0.2 # dex
    rand = norm.rvs(loc=0, scale=0.2, size=len(rvir)) if addScatter else 0.
    log_rvir = np.log10(rvir)
    log_rh = slope * log_rvir + rand + np.log10(normalization)
    return 10**log_rh
```

```python
rvir = np.logspace(-1, 2, 1000)
fig, ax = plt.subplots(dpi=200, figsize=(4,3))
ax.scatter(.015 * rvir, rh_rvir_relation(rvir), s=.2, c='cornflowerblue')
ax.plot(.015 * rvir, rh_rvir_relation(rvir, False), color='red')
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_ylabel(r'$r_{1/2}$ [kpc]')
ax.set_xlabel(r'$0.015 \times R_\mathrm{vir}$ [kpc]')
ax.scatter(1, 1, marker='x', c='purple')
plt.show()
```

```python
rh_rvir_relation([300.], True) / 1.3
```
