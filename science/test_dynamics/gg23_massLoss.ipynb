# setup potential

```python
dm     = gp.NFWPotential(m=1e12, r_s=25., units=galactic)
baryon = gp.PlummerPotential(m=4e10, b=1.6, units=galactic)
pot    = dm + baryon

def getTidalStrength(q):
    hess = pot.hessian(q)
    tidal_tensor = hess - np.multiply((1/3)*np.trace(hess, axis1=0, axis2=1), np.tile(np.identity(3), (hess.shape[2], 1, 1)).T)
    eigs = np.linalg.eigvals(tidal_tensor.T)
    lam = np.max(np.abs(eigs), axis=1)
    return lam
```

```python
# palomar 5 orbital parameters
r_apo = 15.5
r_peri = 6.5
ecc = 0.41

w0 = gd.PhaseSpacePosition(pos=[r_apo, 0., 0.] * u.kpc, vel=[0., 150., 0.] * u.km/u.s)
particle_pot = [None]


# vary these
# time_step = 1e-2*u.Myr
# initial_mass = 4.7e4

####################
time_steps     = [1e-2, 1e-1, 1e1, 1e2] * u.Myr
initial_masses = np.logspace(3, 5, 5)
dmdt = np.zeros((len(time_steps), len(initial_masses)))
dmdt_gg = np.zeros((len(time_steps), len(initial_masses)))

dE = np.zeros((len(time_steps), len(initial_masses)))
####################

for k, time_step in tqdm(enumerate(time_steps)):
    for j, initial_mass in tqdm(enumerate(initial_masses)):
        
        nbody = DirectNBody(w0, particle_pot, external_potential=pot)
        orbits = nbody.integrate_orbit(dt=time_step, t1=0., t2=1*u.Gyr)
        lams = getTidalStrength(orbits[:, 0].xyz)
        dt = (orbits[:, 0].t[1:] - orbits[:, 0].t[:-1]).to(u.Gyr)[0].value
        mass = np.zeros(len(orbits[:, 0].t))
        gg_mass = np.zeros(len(orbits[:, 0].t))
        
        mass[0] = initial_mass
        gg_mass[0] = initial_mass
        
        for i in tqdm(range(len(orbits[:, 0].t)-1)):
            mass[i+1] = mass[i] + getMassLossRate(mass[i], lams[i]).to(1/u.Gyr).value * dt
            gg_mass[i+1] = gg_mass[i] + mdot_gg23(gg_mass[i], lams[i], initial_mass).to(1/u.Gyr).value * dt
        
        _dmdt = np.abs(mass[-1] - initial_mass)
        _dmdt_gg = np.abs(gg_mass[-1] - initial_mass)

        dmdt[k, j] = _dmdt
        dmdt_gg[k, j] = _dmdt_gg
```

```python
fig, ax = plt.subplots(dpi=200, figsize=(4,4))

ax.loglog(initial_masses, 3e-2*initial_masses**(.33), ls='--', c='k', label=r'$\dot{M} \propto M^{0.33}$')

ax.loglog(initial_masses, dmdt[0]/1000, c='cornflowerblue', alpha=1, marker='o', label=r'CG23; $\log_{10} (\Delta t / \mathrm{Myr}) = -2$')
ax.loglog(initial_masses, dmdt[1]/1000, c='cornflowerblue', alpha=.75, marker='o', label=r'$\log_{10} (\Delta t / \mathrm{Myr}) = -1$')
ax.loglog(initial_masses, dmdt[2]/1000, c='cornflowerblue', alpha=.5, marker='o', label=r'$\log_{10} (\Delta t / \mathrm{Myr}) =1$')
ax.loglog(initial_masses, dmdt[3]/1000, c='cornflowerblue', alpha=.25, marker='o', label=r'$\log_{10} (\Delta t / \mathrm{Myr}) =2$')

ax.loglog(initial_masses, dmdt_gg[0]/1000, c='firebrick', marker='^', alpha=1, label='GG23')
ax.loglog(initial_masses, dmdt_gg[1]/1000, c='firebrick', marker='^', alpha=.75)
ax.loglog(initial_masses, dmdt_gg[2]/1000, c='firebrick', marker='^', alpha=.5)
ax.loglog(initial_masses, dmdt_gg[3]/1000, c='firebrick', marker='^', alpha=.25)

ax.set_ylim(10**(-1), 10**(2))
ax.legend()
ax.invert_xaxis()

ax.set_ylabel(r'$\dot{M}$ [M$_\odot$/Myr]')
ax.set_xlabel(r'$M_\mathrm{initial}$ [M$_\odot$]')
plt.show()
```
