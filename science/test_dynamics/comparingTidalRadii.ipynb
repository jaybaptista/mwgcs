```python
sim_count = 4
sim_dirs = glob("/fs/ddn/sdf/group/kipac/g/cosmo/ki21/phil1/simulations/ZoomIns/SymphonyMilkyWay/Halo*")
sims = [Simulation(sim_dir) for sim_dir in tqdm(sim_dirs[:sim_count])]
```

```python
m = 2e5
R_gal = np.logspace(-1, 2, 50)

rts = []
lams = []
oms = []

profiles = []

for sim in tqdm(sims):
    profile = MassProfile(sim, 235, 0)
    _ = profile.fit()
    profiles.append(profile)
    
    # Mr = np.array([profile.mass(R_i) for R_i in R_gal])    
    # slope = profile._ein.analyticSlope(R_gal)
    # r_t_mw = R_gal * np.array(((m/Mr) / (3 - slope))**(1/3)) * u.kpc
    
    r_t_mw = np.array([tidalRadius(R_i, m, profile) for R_i in R_gal], dtype=object) * u.kpc
    lam = (cons.G * m * u.Msun / (r_t_mw)**3).to(u.Gyr**(-2))
    om = ((lam / 3)**(1/2)).to(1/u.Gyr)
        
    rts.append(r_t_mw)
    lams.append(lam)
    oms.append(om)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(11,3), dpi=300)

ax[0].loglog([], [], c='dodgerblue', alpha=.2, label='MW Halos')

for r_t_mw, lam, om in tqdm(zip(rts, lams, oms)):
    ax[0].loglog(R_gal, r_t_mw, c='dodgerblue', alpha=.2)
    ax[1].loglog(R_gal, lam, c='dodgerblue', alpha=.2)
    ax[2].loglog(R_gal, om, c='dodgerblue', alpha=.2)

# ax[2].loglog(R_gal, 4*om, c='mediumseagreen', alpha=.5, ls='--')

ax[0].set_title('Tidal Radius')
ax[1].set_title('Tidal Strength')
ax[2].set_title('Tidal Frequency')

ax[0].set_ylabel(r'$r_\mathrm{tidal}$ [kpc]')
ax[1].set_ylabel(r'$\lambda$ [Gyr$^{-2}$]')
ax[2].set_ylabel(r'$\Omega_\mathrm{tid}$ [Gyr$^{-1}$]')

ax[0].plot([3, 3], [0.02, 0.05], marker='_', color='lightcoral', label='CG22')
ax[1].scatter(3, 1e4, marker='x', c='lightcoral', label='CG22')
ax[2].scatter(3, 230, marker='x', c='lightcoral', label='CG22')

ax[0].legend()

fig.suptitle(r'$\log_{10} (M_\mathrm{GC} / M_\odot) = 5$')

fig.text(0.5, 0, 'r [kpc]', ha='center')

fig.tight_layout()
plt.show()
```

```python
lams_r = []
oms_r = []
rts_r = []

lams_r_vdb = []
oms_r_vdb = []
rts_r_vdb = []

lams_r_ae = []
oms_r_ae = []
rts_r_ae = []

radii = np.logspace(-1, 2, 50)
m = 2e5 * u.Msun

for i in np.arange(len(rs)):
    _lams = []
    _oms = []
    _rts = []
    
    _lams_vdb = []
    _oms_vdb = []
    _rts_vdb = []
    
    _lams_ae = []
    _oms_ae = []
    _rts_ae = []
    
    for r_i in radii:

        tt = nfws[i].getTidalTensor(r_i*u.kpc)
        tt_ae = profiles[i]._ein.analyticTidalTensor(r_i*u.kpc)        
        
        # tidal parameters from analytic NFW tidal tensor derivation
        lam = (np.max(np.abs(np.linalg.eig(tt)[0]))).to(1/u.Gyr**2).value
        om = ((lam / 3)**(1/2))
        rt = ((cons.G * m / (lam / u.Gyr**2))**(1/3)).to(u.kpc).value
        
        # tidal parameters from analytic Einasto tidal tensor derivation
        lam_ae = (np.max(np.abs(np.linalg.eig(tt_ae)[0]))).to(1/u.Gyr**2).value
        om_ae = ((lam_ae / 3)**(1/2))
        rt_ae = ((cons.G * m / (lam_ae / u.Gyr**2))**(1/3)).to(u.kpc).value
        
        # tidal parameters from vdB+18 approximation
        rt_vdb = tidalRadius(r_i, m, nfws[i])
        lam_vdb = getLambda(m, rt_vdb)
        om_vdb = ((lam_vdb / 3)**(1/2)).to(1/u.Gyr).value
        
        _lams.append(lam)
        _oms.append(om)
        _rts.append(rt)
        
        _lams_vdb.append(lam_vdb)
        _oms_vdb.append(om_vdb)
        _rts_vdb.append(rt_vdb)
        
        _lams_ae.append(lam_ae)
        _oms_ae.append(om_ae)
        _rts_ae.append(rt_ae)
        
    lams_r.append(_lams)
    oms_r.append(_oms)
    rts_r.append(_rts)
    
    lams_r_vdb.append(_lams_vdb)
    oms_r_vdb.append(_oms_vdb)
    rts_r_vdb.append(_rts_vdb)
    
    lams_r_ae.append(_lams_ae)
    oms_r_ae.append(_oms_ae)
    rts_r_ae.append(_rts_ae)
    
lams_r = np.array(lams_r)
oms_r = np.array(oms_r)
rts_r = np.array(rts_r)
```
