```python
# w_sat = PhaseSpacePosition([10., 3., 2.] * u.kpc, vel = [3., 20., 20.] * u.km / u.s)

r_apo = 15.5
dm     = gp.NFWPotential(m=1e12, r_s=25., units=galactic)
baryon = gp.PlummerPotential(m=4e10, b=1.6, units=galactic)
# baryon = gp.MiyamotoNagaiPotential(m=1e10, a=3, b=2, units=galactic)
pot    = dm + baryon


dt=1e-2*u.Myr
q0 = [r_apo, 0., 2.] * u.kpc
v_init = pot.circular_velocity(q0).value[0] * .5
v_i = [0., v_init, 0.] * u.km / u.s
w_sat = gd.PhaseSpacePosition(pos=q0, vel=v_i)


end_time = 6000 * u.Myr

particle_pot = [None]
nbody = DirectNBody(w_sat, particle_pot, external_potential=pot)
T = 2*np.pi*r_apo * u.kpc / pot.circular_velocity(q0) 
orb_sat = nbody.integrate_orbit(dt=dt, t1=0, t2=end_time)

r_apo = 15.5 * u.kpc
```

```python
def getTidalStrength(q):
    hess = pot.hessian(q)
    tidal_tensor = hess - np.multiply((1/3)*np.trace(hess, axis1=0, axis2=1), np.tile(np.identity(3), (hess.shape[2], 1, 1)).T)
    eigs = np.linalg.eigvals(tidal_tensor.T)
    lam = np.max(np.abs(eigs), axis=1)
    return lam

test_radii     = np.logspace(-2, 2, 200)
test_qs        = np.vstack([np.zeros(len(test_radii)), test_radii, np.zeros(len(test_radii))]).T * u.kpc
test_strengths = np.log10([getTidalStrength(q).to(u.Gyr**(-2)).value[0] for q in test_qs])
interp_log_strength = interp1d(np.log10(test_radii), test_strengths, kind='linear', fill_value='extrapolate')

m0 = 3e4
r = np.sqrt(np.sum(orb_sat[:, 0].xyz**2, axis=0))
l = 10**(interp_log_strength(np.log10(r.value))) / u.Gyr**2
masses = CMassLoss(m0, l, orb_sat.shape[0], dt)

r_peri = orb_sat.pericenter()[0]
```

```python
rounded_mass_loss = np.around(np.cumsum(np.abs(np.gradient(masses))))
```

```python
pos_ejs = []
vel_ejs = []
t0s = []
for k in tqdm(range(orb_sat.shape[0]-1)):
    # dM = rounded_mass_loss[k+1] - rounded_mass_loss[k]
    if k % 10000 == 0: # hack rn, replace this with the distribution function along the orbit
    # if (dM >= 1.) and (k % 10000) :
        w_sat = orb_sat[k, 0].represent_as('cylindrical')
        w_ej = w_ejecta(w_sat, masses[k] * u.Msun, pot, r_apo, r_peri)
        
        t0 = orb_sat.t[k]
        t0s.append(t0)
        t0s.append(t0)

        _p = w_ej.pos.to_cartesian().get_xyz().to(u.kpc)
        _q = w_ej.vel.get_d_xyz().to(u.km / u.s)
        
        pos_ejs.append(_p[:, 0])
        pos_ejs.append(_p[:, 1])
        vel_ejs.append(_q[:, 0])
        vel_ejs.append(_q[:, 1])
        
        # orb_part = nbody.integrate_orbit(dt=dt, t1=t0, t2=T)

pos_ejs = u.Quantity(pos_ejs).T[0]
vel_ejs = u.Quantity(vel_ejs).T[0]
```
