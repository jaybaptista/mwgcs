# Ramos-Montenegro Expression for Analytic Einasto Potential

**NOTE**: $R_s = r_{-2}$

```python
def analyticEinastoDensity():
    pass

def analyticNFWDensity():
    pass

def analyticEinastoPotential(r, m=1e12 * u.Msun, alpha=.18, Rs=25*u.kpc):
        # see eqn 19 from https://arxiv.org/abs/1202.5242
        dn = 2 / alpha
        n = 1 / alpha
        
        s = (dn)**n * (r / Rs).decompose().value
        h = Rs / (2*n)**n
        t1 = gammaincc(3*n, s**(1/n))
        t2 = gammaincc(2*n, s**(1/n)) * s * (gamma(2*n) / gamma(3*n))
        # see 
        return -1 * (cons.G * m / (h * s)).to((u.km / u.s)**2) * (1 - t1 + t2)
    
def analyticNFWPotential(r, m=1e12 * u.Msun, conc=11., Rs=25*u.kpc):
        # based on NFW profile wikipedia page scaling relations :/ 
        rho0 = m / (4 * np.pi * Rs**3 * (np.log(1+conc) - (conc/(1+conc))))
        coeff1 = 4 * np.pi * cons.G
        return -1*(coeff1 * rho0 * (Rs**3 / r)).to((u.km / u.s)**2) * np.log(1 + (r / Rs))
```

```python
Rs = 25

target_radius = 3.
target_tidal_frequency = (250 / u.Gyr).to(u.s**(-1))
target_eigenvalue = (2/3 * target_tidal_frequency**2).to(u.s**(-2)).value
```

```python
fig, ax = plt.subplots(1, 3, dpi=300, figsize=(7,2))
radii = np.logspace(-2, 6, 1000)
einastoPotential = [analyticEinastoPotential(_r * u.kpc).value for _r in radii]
NFWPotential = [analyticNFWPotential(_r * u.kpc).value for _r in radii]
einastoPotential_r = np.gradient(einastoPotential, edge_order=2)
NFWPotential_r = np.gradient(NFWPotential, edge_order=2)
einastoPotential_rr = np.gradient(einastoPotential_r, edge_order=2)
NFWPotential_rr = np.gradient(NFWPotential_r, edge_order=2)

ax[0].plot(np.log10(radii/Rs), einastoPotential, 'b', label='Einasto')
ax[0].plot(np.log10(radii/Rs), NFWPotential, 'r', label='NFW')
ax[0].set_xlabel(r'$\log (r/R_s)$')
ax[0].set_ylabel(r'$\Phi$ [$\mathrm{km^2 / s^2}$]')
# ax[0].set_xscale('log')
ax[0].set_yscale('symlog')
ax[0].legend()
ax[0].axhline(0, ls='--', c='k', alpha=.2)

ax[1].plot(radii/Rs, einastoPotential_r, 'b')
ax[1].plot(radii/Rs, NFWPotential_r, 'r')
ax[1].set_xlabel(r'$r/R_s$')
ax[1].set_xscale('log')
ax[1].set_yscale('log')
ax[1].set_ylabel(r'$\frac{\partial \Phi}{\partial r}$ [$\mathrm{km / s^2}$]')

ax[2].plot(radii/Rs, einastoPotential_rr, 'b')
ax[2].plot(radii/Rs, NFWPotential_rr, 'r')
ax[2].set_xlabel(r'$r/R_s$')
ax[2].set_xscale('log')
ax[2].scatter(target_radius, target_eigenvalue, c='k', marker='x')
ax[2].set_ylabel(r'$\frac{\partial^2 \Phi}{\partial r^2}$ [$\mathrm{1 / s^2}$]')

fig.tight_layout()
plt.show()
```

# Einasto Mass

```python
def EinastoMassProfile(r, m=1e12 * u.Msun, alpha=.18, Rs=25*u.kpc):
    dn = 2 / alpha
    n = 1 / alpha

    s = (dn)**n * (r / Rs).decompose().value
    h = Rs / (2*n)**n 
    
    return m * (1 - gammaincc(3*n, s**(1/n)))
```
