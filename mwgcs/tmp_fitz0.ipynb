```python
params = symlib.simulation_parameters(sim_dir)
scale_factors = symlib.scale_factors(sim_dir)
mp = params['mp'] / params['h100']
particle_class = symlib.Particles(sim_dir)

snapshot = 235
halo_id = 0

particles = particle_class.read(snapshot, mode='all')
q = particles[halo_id]['x']
```

```python
q_sh = rs[halo_id, snapshot]['x']
r_conv = 3 * (params['eps']/params['h100']) * scale_factors[snapshot]
profile = SymphonyHaloProfile(q, q_sh, mp, rs[halo_id, snapshot]['rvir'])
```

```python
dist = np.sqrt(np.sum((q - q_sh)**2, axis=1))
```

```python
fp = profile.fit(r_conv, bins=40)
```

```python
rvir, cvir, mvir = rs[halo_id, snapshot]['rvir'], rs[halo_id, snapshot]['cvir'], rs[halo_id, snapshot]['m']

def get_nfw_scale_density(r, r_vir, c, M_vir):
    # Scale radius
    r_s = r_vir / c
    
    # Define the characteristic density rho_0
    rho_0 = M_vir / (4 * np.pi * r_s**3 * (np.log(1 + c) - (c / (1 + c))))
    return rho_0
def nfw_density_profile(r, r_vir, c, M_vir):
    
    # Scale radius
    r_s = r_vir / c
    # A_NFW = (np.log(1 + c) - (c / (1 + c)))
    # M_s = M_vir / A_NFW
    
    # Define the characteristic density rho_0
    rho_0 = M_vir / (4 * np.pi * r_s**3 * (np.log(1 + c) - (c / (1 + c))))
    
    # Calculate the density profile
    rho_r = rho_0 / ((r / r_s) * (1 + r / r_s)**2)
    
    return rho_r
```

```python
einasto = Einasto(fp['alpha'], fp['Rs'], fp['logScaleDensity'])
A_NFW = np.log(1+cvir) - cvir/(1+cvir)
nfw = NFW(mvir, rvir, cvir)
# nfw     = NFWPotential(mvir / A_NFW * u.Msun, rvir / cvir * u.kpc, units=galactic)
```

```python
bin_radius, bin_density = profile.getDensityProfile(40)
# einasto_density = 10**profile.einasto_log_density(bin_radius, fp['alpha'], fp['Rs'], fp['logScaleDensity'])
einasto_density = einasto.density(bin_radius)
# nfw_density = nfw.density([bin_radius, 0*bin_radius, 0*bin_radius])
nfw_density = nfw.density(bin_radius)
```

```python
fig, ax = plt.subplots(dpi=200)
ax.loglog(bin_radius, bin_density, label='binned data', c='blue', marker='o')
ax.loglog(bin_radius, einasto_density, ls='--', c='r', label='einasto fit')
ax.loglog(bin_radius, nfw_density, label='nfw profile', c='orange')
ax.set_ylabel(r'$\rho(<r)$')
ax.set_xlabel('$r$ [kpc]')
ax.text(.15, .3, r'$\frac{r_\mathrm{vir}}{c_\mathrm{vir}} \sim 23.58$ kpc', ha='left', transform=ax.transAxes)

ax.text(.15, .2, r'$\log_{10} \left(\frac{\rho_\mathrm{nfw}}{M_\odot / \mathrm{\, kpc^3}}\right) \sim 6.0$', ha='left', transform=ax.transAxes)
ax.legend()
plt.show()

# how much variability
# look at fluctuations over time
```

```python
radii = np.logspace(-3, 2, 1000)
menc = np.array([profile.menc(r_i) for r_i in radii])

_g_v = 4.300917270036279e-06 #((u.kpc / u.Msun) * (u.km/u.s)**2)

vcirc = np.sqrt(_g_v * menc / radii)

_g_accel = 1.3938323614347172e-22 # ((u.kpc**2 / u.Msun) * (u.km/u.s**2))
accel = _g_accel * menc / radii**2

# einasto acceleration

ein_accel = np.array([einasto.acceleration(r_i) for r_i in radii])
nfw_accel = np.abs(np.array([nfw.acceleration(r_i) for r_i in radii]))
```
